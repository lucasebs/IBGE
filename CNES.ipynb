```python
import pandas as pd
```

```python
arq_municipios = 'tbMunicipio201807'
df_municipios = pd.read_csv('CNES/BASE_DE_DADOS_CNES_201807/' + arq_municipios + '.csv', dtype={'user_id': int}  ,sep=';', low_memory=False)
df_municipios = df_municipios[df_municipios['CO_SIGLA_ESTADO'] == 'RN']
df_municipios = df_municipios[df_municipios['NO_MUNICIPIO'] == 'NATAL']
df_municipios
```

```python
estabelecimentos = 'tbEstabelecimento201807'
df_estabelecimentos = pd.read_csv('CNES/BASE_DE_DADOS_CNES_201807/' + estabelecimentos + '.csv', dtype={'user_id': int}  ,sep=';', low_memory=False)
df_estabelecimentos = df_estabelecimentos[df_estabelecimentos['CO_ESTADO_GESTOR'] == 24]
df_estabelecimentos = df_estabelecimentos[df_estabelecimentos['CO_MUNICIPIO_GESTOR'] == 240810]

for i in range(1,15):
    df_estabelecimentos = df_estabelecimentos[df_estabelecimentos['CO_MOTIVO_DESAB'] != i ]

df_estabelecimentos.head()
# df_estabelecimentos
```

```python
df_estabelecimentos2 = df_estabelecimentos.groupby('NO_BAIRRO').NO_BAIRRO.count().to_dict()
```

```python
r_sul = ['CANDELARIA', 'CAPIM MACIO' , 'CIDADE JARDIM', 'LAAGOA NOVA',
       'LAGOA', 'LAGOA NAVA', 'LAGOA NOVA', 'LAGOA NOVAQ',
       'LAGOA NOVQA', 'EDF LAGOA CENTER', 'MIRASSOL', 'MORRO BRANCO',
       'NATAL', 'NEOPOLIS', 'NNOVA DESCOBERTA', 'NOVA DESCOBERTA',
       'PIRANGI', 'PITIMBU', 'CIDADE SATELITE', 'PONTA NEGRA',
       'POTILANDIA']

r_oeste = ['BAIXO DIX SEPT ROSAD', 'BOM PASTOR', 'CIDADE DA ESPERANCA',
         'CIDADE DA ESPEREANCA', 'DIX SEPT ROSADO' , 'DIXSEPT ROSADO',
         'DIXSEPTROSADO', 'FELIPE CAMARAO', 'GUARAPES', 'NAZARE',
         'NORDESTE', 'PLANALTO', 'QUINTAS', 'CIDADE NOVA', 'NOVA CIDADE']

r_norte = ['IGAPO', 'LAGOA AZUL', 'NOSSA S DA APRESENTA', 
         'NOSSA SENHORA DA APR', 'NOSSA SRA DA APRESEN',
         'N S APRESENTACAO', 'N S DA APRESENTACAO',
         'NOVA NATAL', 'PAJUCARA', 'POTENGI', 'REDINHA',
         'SANTAREM', 'VALE DOURADO']

r_leste = ['ALECRIM', 'AREIA PRETA', 'BARO VERMELHO', 'BARRO VERMELHO',
         'BAIRRO VERMELHO', 'BRRO VERMELHO', 'BRASILIA TEIMOSA',
         'CIDADE ALTA', 'CENTRO', 'LAGOA SECA', 'MAE LUIZA', 
         'PETEROPOLIS', 'PETREOPOLIS', 'PETROLPOLIS', 'PETROPILIS',
         'PETROPOIS', 'PETROPOLIS' , 'PETRPOLIS', 'RIBEIRA', 'ROCAS',
         'TIEROL', 'TIOROL', 'TIREOL', 'TIROL' , 'RITOL', 'AFONSO PENA',
         'CIDADE DA CRIANCA']
```

```python
leste = 0
norte = 0 
oeste = 0
sul = 0

for bairro in df_estabelecimentos2.keys():
    if bairro in r_leste:
        leste += df_estabelecimentos2[bairro]
    elif bairro in r_oeste:
        oeste += df_estabelecimentos2[bairro]
    elif bairro in r_sul:
        sul += df_estabelecimentos2[bairro]
    elif bairro in r_norte:
        norte += df_estabelecimentos2[bairro]
```

```python
print('Leste: ' + str(leste) + '\nNorte: ' + str(norte) +
      '\nOeste: ' + str(oeste) + '\nSul: ' + str(sul))
```
